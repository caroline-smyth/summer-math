```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn
seaborn.set()

import wrds
WRDS_LOGIN = 'cssmyth'
```

```python
conn = wrds.Connection(wrds_username=WRDS_LOGIN)
```

```python
wrds_names_query = conn.get_table(library='crsp', table='wrds_names_query') # get v2 table, which has secinfostartdt and enddt 
```

```python
msf = conn.get_table(library='crsp', table='msf') # msf table, which has firm weights
```

```python
# compute weight for all CRSP securities per month
msf['mktcap'] = msf['prc'].abs() * msf['shrout']
msf['total_mktcap'] = msf.groupby('date')['mktcap'].transform('sum')
msf['weight'] = msf['mktcap'] / msf['total_mktcap']

msf['pctile'] = (
    msf.groupby('date')['weight'].rank(pct=True, ascending=False)
)
```

```python
# make date a datetime type

msf['date'] = pd.to_datetime(msf['date'])

# start of security data and end of security data
wrds_names_query['securitybegdt'] = pd.to_datetime(wrds_names_query['securitybegdt'])
wrds_names_query['securityenddt'] = pd.to_datetime(wrds_names_query['securityenddt'])

# turn dates into month period for joins
msf['month']  = msf['date'].dt.to_period('M')
wrds_names_query['beg_month'] = wrds_names_query['securitybegdt'].dt.to_period('M')
wrds_names_query['end_month'] = wrds_names_query['securityenddt'].dt.to_period('M')

# anchor months for price/rank lookups
wrds_names_query['beg_anchor'] = wrds_names_query['beg_month'] - 1
wrds_names_query['end_anchor'] = wrds_names_query['end_month'] - 1
```

```python
# first and last month in msf universe
first_month = msf['month'].min()
last_month  = msf['month'].max()

#msf_cols = ['permno', 'month', 'pctile', 'weight', 'mktcap', 'total_mktcap']

#ent_keys = wrds_names_query.loc[wrds_names_query['beg_month'].notna(), ['permno','beg_month','beg_anchor']].copy()
#ent_keys = ent_keys[ent_keys['beg_anchor'] >= first_month]

drop_months = [
    pd.Period('1962-07', 'M'),
    pd.Period('1972-12', 'M'),
    pd.Period('2006-03', 'M'),
]


ent_keys['next_month'] = ent_keys['beg_month'] + 1  # Period[M] arithmetic

# entrances 
df_ent = pd.merge(
    msf[['permno','month','pctile','weight','mktcap','total_mktcap']],
    ent_keys,
    left_on  = ['permno','month'],
    right_on = ['permno','next_month'],
    how      = 'inner'
)

df_ent['event'] = 'entrance'
df_ent = df_ent[df_ent['month'] != first_month]
df_ent = df_ent[~df_ent['month'].isin(drop_months)]
df_ent = df_ent.assign(event_month=df_ent['month'])

exit_keys = wrds_names_query.loc[
    wrds_names_query['end_month'].notna(),
    ['permno','end_month']
].copy()

exit_keys['prev_month'] = exit_keys['end_month'] - 1  # Period[M] arithmetic


# exits 
df_exit = pd.merge(
    msf[['permno','month','pctile','weight','mktcap','total_mktcap']],
    exit_keys, # wrds_names_query[['permno','end_month']] 
    left_on  = ['permno','month'],
    right_on = ['permno','prev_month'],
    how      = 'inner'
)

df_exit = df_exit[~df_exit['month'].isin(drop_months)]

df_exit['event'] = 'exit'
df_exit = df_exit[df_exit['month'] != last_month]
df_exit = df_exit[~df_exit['month'].isin(drop_months)]

df_exit = (df_exit
           .rename(columns={'end_month':'event_month'})
           .drop(columns=['prev_month'])
          )

# collapse to one entrance (earliest) and exit (latest) per permno
df_ent = (df_ent
          .sort_values('month')                    
          .drop_duplicates(subset='permno',     
                           keep='first')          
         )
df_exit = (df_exit
           .sort_values('month')                   
           .drop_duplicates(subset='permno',     
                            keep='last')          
          )

# combine
df_events = pd.concat([df_ent, df_exit], ignore_index=True)

# rename if downstream expects quantile but p sure it doesn't
df_events.rename(columns={'pctile':'quantile'}, inplace=True)
```

```python
# fit beta distributions with data adjustment
def fit_beta_distribution(data, event_type):
    """
    Fit beta distribution to quantile data
    Returns parameters and goodness-of-fit statistics
    """
    # handle boundary values
    clean_data = data.dropna()
    #clean_data = data
    
    # Beta distribution requires values strictly between 0 and 1
    # apply small adjustment to boundary values
    epsilon = 1e-6
    clean_data = clean_data.copy()
    clean_data[clean_data <= 0] = epsilon
    clean_data[clean_data >= 1] = 1 - epsilon
    
    print(f"\n{event_type} Data Summary:")
    print(f"Original range: [{data.min():.6f}, {data.max():.6f}]")
    print(f"Adjusted range: [{clean_data.min():.6f}, {clean_data.max():.6f}]")
    print(f"Number of boundary adjustments: {((data <= 0) | (data >= 1)).sum()}")

    # fit beta distribution using method of moments and MLE
    alpha_hat, beta_hat, loc, scale = stats.beta.fit(clean_data, floc=0, fscale=1)
    
    # calc goodness of fit using adjusted data
    ks_stat, ks_pvalue = stats.kstest(clean_data, lambda x: stats.beta.cdf(x, alpha_hat, beta_hat))
    
    # calc aic, bic
    log_likelihood = np.sum(stats.beta.logpdf(clean_data, alpha_hat, beta_hat))
    n = len(clean_data)
    aic = 2 * 2 - 2 * log_likelihood  # 2 parameters (alpha, beta)
    bic = np.log(n) * 2 - 2 * log_likelihood
    
    results = {
        'alpha': alpha_hat,
        'beta': beta_hat,
        'ks_statistic': ks_stat,
        'ks_pvalue': ks_pvalue,
        'aic': aic,
        'bic': bic,
        'log_likelihood': log_likelihood,
        'n_observations': n,
        'adjusted_data': clean_data  # Store the adjusted data for later use
    }

    print(f"\n{event_type} Beta Distribution Parameters:")
    print(f"Alpha: {alpha_hat:.4f}")
    print(f"Beta: {beta_hat:.4f}")
    print(f"KS Test p-value: {ks_pvalue:.4f}")
    print(f"AIC: {aic:.2f}")
    print(f"BIC: {bic:.2f}")
    
    return results

# fit distributions
entrance_params = fit_beta_distribution(entrance_data, "ENTRANCE")
exit_params = fit_beta_distribution(exit_data, "EXIT")
```

```python
# calculate probabilities by market rank deciles
def calculate_entry_exit_probabilities(entrance_params, exit_params, quantile_range=None):
    """
    Calculate probabilities of entry/exit for different market rank ranges
    """
    if quantile_range is None:
        quantile_range = np.linspace(0, 1, 11)  # Deciles
    
    results = []
    
    for i in range(len(quantile_range)-1):
        lower = quantile_range[i]
        upper = quantile_range[i+1]
        
        # calculate probabilities using CDF
        entrance_prob = (stats.beta.cdf(upper, entrance_params['alpha'], entrance_params['beta']) - 
                        stats.beta.cdf(lower, entrance_params['alpha'], entrance_params['beta']))
        
        exit_prob = (stats.beta.cdf(upper, exit_params['alpha'], exit_params['beta']) - 
                    stats.beta.cdf(lower, exit_params['alpha'], exit_params['beta']))
        
        results.append({
            'quantile_range': f"{lower:.1f}-{upper:.1f}",
            'entrance_probability': entrance_prob,
            'exit_probability': exit_prob,
            'ratio_exit_to_entrance': exit_prob / entrance_prob if entrance_prob > 0 else np.inf
        })
    
    return pd.DataFrame(results)

# calculate probabilities
prob_df = calculate_entry_exit_probabilities(entrance_params, exit_params)
print("PROBABILITY ANALYSIS BY MARKET RANK DECILE:")
print(prob_df.round(4))
```

```python
# visualize probability analysis
fig, ax = plt.subplots(figsize=(12, 6))

# Probability comparison by decile
deciles = range(len(prob_df))
width = 0.35

ax.bar([x - width/2 for x in deciles], prob_df['entrance_probability'], 
           width, label='Entrance Probability', alpha=0.7, color='blue')
ax.bar([x + width/2 for x in deciles], prob_df['exit_probability'], 
           width, label='Exit Probability', alpha=0.7, color='red')
ax.set_xlabel('Market Rank Decile')
ax.set_ylabel('Probability')
ax.set_title('Entry vs Exit Probabilities by Market Rank Decile')
ax.set_xticks(deciles)
ax.set_xticklabels(prob_df['quantile_range'], rotation=45)
ax.legend()
ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
# simulation horizon, in month units bc lambdas are per month
t = 0.0
i = 0
events = []  # to record (type, time, x)

# main loop - one Exp clock per particle
#while t < T_sim and len(particles) > 0:
while i < 10000:  
    # compute each particles total intensity
    i +=1 # counting loop iterations to make sure it doesn't just stop at 1
    rates = _total(particles)         # vector of _i
    # pull waiting times ~ Exp(_i)
    taus = np.random.exponential(1.0 / rates)
    
    # find next event
    i_star = np.argmin(taus)
    dt = taus[i_star]
    t += dt
    x_i = particles[i_star]
    
    # pick birth vs death
    if np.random.rand() < _birth(x_i) / _total(x_i):
        # birth: add a new particle at the same percentile
        particles = np.append(particles, x_i)
        particles = np.clip(particles, 1e-6, 1-1e-6)
        events.append(('birth', t, x_i))
    else:
        # death: remove that particle
        events.append(('death', t, x_i))
        particles = np.delete(particles, i_star)
#print("Loop iterations:", i)
#print("Events recorded:", len(events))
```

```python
import numpy as np
import pandas as pd
from scipy import stats

# --- split simulated samples ---
sim_birth = df_sim.loc[df_sim['event']=='birth', 'percentile'].to_numpy()
sim_death = df_sim.loc[df_sim['event']=='death', 'percentile'].to_numpy()

def one_sample_gof_beta(sample, alpha, beta, label, m_bins=10, seed=0):
    """One-sample GOF vs Beta(alpha, beta):
       KS, Cramrvon Mises, binned chi-square (equal-probability bins), Wasserstein-1."""
    sample = np.asarray(sample)
    sample = sample[(sample > 0) & (sample < 1)]
    out = {'n': sample.size}
    if sample.size < 10:
        out['note'] = f'[{label}] too few samples for reliable GOF'
        return out
    
    # KS and CvM against the theoretical Beta CDF
    ks   = stats.kstest(sample, 'beta', args=(alpha, beta))
    cvm  = stats.cramervonmises(sample, stats.beta.cdf, args=(alpha, beta))
    
    # Chi-square on equal-probability bins (using Beta quantiles)
    q = np.linspace(0, 1, m_bins + 1)
    bins = stats.beta.ppf(q, alpha, beta)
    bins[0], bins[-1] = 0.0, 1.0  # guard against -inf/inf edges
    hist, _ = np.histogram(sample, bins=bins)
    expected = np.full_like(hist, sample.size / m_bins, dtype=float)
    chi2 = stats.chisquare(hist, f_exp=expected)
    
    # Wasserstein distance vs a synthetic Beta sample (effect size, no p-value)
    rng = np.random.default_rng(seed)
    ref = stats.beta.rvs(alpha, beta, size=min(5000, max(1000, 2*sample.size)), random_state=rng)
    w1  = stats.wasserstein_distance(sample, ref)
    
    return {
        'n': sample.size,
        'ks_stat': ks.statistic,   'ks_p': ks.pvalue,
        'cvm_stat': cvm.statistic, 'cvm_p': cvm.pvalue,
        'chi2_stat': chi2.statistic, 'chi2_p': chi2.pvalue,
        'w1': w1
    }

print('One-sample GOF vs target Beta')
birth_gof = one_sample_gof_beta(sim_birth, _b, _b, 'birth')
death_gof = one_sample_gof_beta(sim_death, _d, _d, 'death')
print('Birth:', birth_gof)
print('Death:', death_gof)

# --- Calibration test for birth vs death choices ---
# For each recorded event with percentile x_j, predicted P(birth) = _birth(x_j) / _total(x_j).
xj = df_sim['percentile'].to_numpy()
pj = stats.beta.pdf(xj, _b, _b)
qj = stats.beta.pdf(xj, _d, _d)
p_hat = (birth_occurrence * pj) / (birth_occurrence * pj + death_occurrence * qj)  # predicted birth prob per event
y = (df_sim['event'].to_numpy() == 'birth').astype(float)  # 1 if birth, else 0

# Score test (approx N(0,1) under null):
S = np.sum(y - p_hat)
V = np.sum(p_hat * (1.0 - p_hat))
z = S / np.sqrt(V) if V > 0 else np.nan
pval_cal = 2 * stats.norm.sf(np.abs(z))
print('\nBernoulli calibration across events')
print(f'Z = {z:.3f}, p = {pval_cal:.4g}  (H0: events are birth with predicted prob p_hat(x))')

# --- Inter-arrival time sampler check ---
# Minimal edits inside your loop:
#   * before picking i_star, record `Lambda_sum = rates.sum()`
#   * after computing `dt`, append to lists: `lambdas.append(Lambda_sum)` and `dts.append(dt)`
# Then this block will work:

if 'dts' in globals() and 'lambdas' in globals() and len(dts) == len(lambdas) and len(dts) > 10:
    U = 1.0 - np.exp(-np.array(lambdas) * np.array(dts))  # should be Uniform(0,1)
    ks_u = stats.kstest(U, 'uniform')
    print('\n=== Event-time sampler check ===')
    print(f'Uniform-transform KS: stat={ks_u.statistic:.4f}, p={ks_u.pvalue:.4g}')
else:
    print('\n[Note] To test event times, record per-step `lambdas` and `dts` inside the loop as noted above.]')
```

```python
import numpy as np
import pandas as pd
from scipy import stats

# --- containers for new tests ---
pit_birth, pit_death = [], []              # hazard-weighted PIT values
lambdas, dts = [], []                      # for event-time transform

# dynamic expected vs observed in percentile bins
m_bins = 12
bins = np.linspace(0.0, 1.0, m_bins+1)
exp_birth_bins = np.zeros(m_bins)
obs_birth_bins = np.zeros(m_bins)
exp_death_bins = np.zeros(m_bins)
obs_death_bins = np.zeros(m_bins)

rng = np.random.default_rng(0)

# === your loop, with a few NEW lines ===
t = 0.0
i = 0
events = []
while i < 1000 and len(particles) > 0:
    i += 1
    # current per-particle rates
    rates_tot   = _total(particles)
    rates_birth = _birth(particles)
    rates_death = _death(particles)

    # event-time bookkeeping (for rescaling test)
    Lambda_sum = rates_tot.sum()

    # draw per-particle waiting times
    taus = np.random.exponential(1.0 / rates_tot)
    i_star = np.argmin(taus)
    dt = taus[i_star]
    t += dt
    x_i = particles[i_star]

    # record for event-time transform (done once per step)
    lambdas.append(Lambda_sum)
    dts.append(dt)

    # choose event type at the chosen location
    is_birth = rng.random() < (rates_birth[i_star] / rates_tot[i_star])

    # ---- Hazard-weighted PIT + dynamic-bin expectation (BIRTH) ----
    if is_birth:
        w = rates_birth
        denom = w.sum()
        # PIT with randomized tie handling (DunnSmyth)
        lt_mask = particles < x_i
        eq_mask = particles == x_i
        num = w[lt_mask].sum()
        ties = w[eq_mask].sum()
        u = rng.random()
        pit = (num + u * ties) / denom
        pit_birth.append(pit)

        # expected bin contributions this step (conditional on birth)
        step_weights = np.bincount(np.digitize(particles, bins) - 1,
                                   weights=w / denom, minlength=m_bins)
        exp_birth_bins += step_weights
        # observed bin
        b_idx = np.digitize(x_i, bins) - 1
        if 0 <= b_idx < m_bins:
            obs_birth_bins[b_idx] += 1

        # apply birth update (your current rule: clone at same x)
        particles = np.append(particles, x_i)
        particles = np.clip(particles, 1e-6, 1-1e-6)
        events.append(('birth', t, x_i))

    # ---- Hazard-weighted PIT + dynamic-bin expectation (DEATH) ----
    else:
        w = rates_death
        denom = w.sum()
        lt_mask = particles < x_i
        eq_mask = particles == x_i
        num = w[lt_mask].sum()
        ties = w[eq_mask].sum()
        u = rng.random()
        pit = (num + u * ties) / denom
        pit_death.append(pit)

        step_weights = np.bincount(np.digitize(particles, bins) - 1,
                                   weights=w / denom, minlength=m_bins)
        exp_death_bins += step_weights

        b_idx = np.digitize(x_i, bins) - 1
        if 0 <= b_idx < m_bins:
            obs_death_bins[b_idx] += 1

        # apply death update
        particles = np.delete(particles, i_star)
        events.append(('death', t, x_i))

# === TESTS ===

print('\n=== Hazard-weighted PIT (Uniform(0,1)) ===')
if len(pit_birth) >= 20:
    ks_birth = stats.kstest(pit_birth, 'uniform')
    print(f'Birth PIT  : KS stat={ks_birth.statistic:.4f}, p={ks_birth.pvalue:.4g}, n={len(pit_birth)}')
else:
    print('Birth PIT  : not enough events')

if len(pit_death) >= 20:
    ks_death = stats.kstest(pit_death, 'uniform')
    print(f'Death PIT  : KS stat={ks_death.statistic:.4f}, p={ks_death.pvalue:.4g}, n={len(pit_death)}')
else:
    print('Death PIT  : not enough events')

print('\n=== Dynamic binwise chi-square (counts vs expected) ===')
def chi2_test(obs, exp, label):
    mask = (exp > 1e-12)  # avoid zero-expected bins
    if mask.sum() >= 3:
        chi2 = stats.chisquare(obs[mask], f_exp=exp[mask])
        print(f'{label:>6}: chi2={chi2.statistic:.2f}, df={mask.sum()-1}, p={chi2.pvalue:.4g}')
    else:
        print(f'{label:>6}: insufficient expected mass')

chi2_test(obs_birth_bins, exp_birth_bins, 'Birth')
chi2_test(obs_death_bins, exp_death_bins, 'Death')

print('\n=== Event-time rescaling (Uniform(0,1)) ===')
if len(dts) >= 20:
    U = 1.0 - np.exp(-np.array(lambdas) * np.array(dts))
    ks_u = stats.kstest(U, 'uniform')
    print(f'Uniformized inter-arrivals: KS stat={ks_u.statistic:.4f}, p={ks_u.pvalue:.4g}, n={len(U)}')
else:
    print('Not enough events to test timing.')
```
