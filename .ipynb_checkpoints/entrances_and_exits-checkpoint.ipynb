```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn; seaborn.set()
import warnings; warnings.filterwarnings('ignore')
from pathlib import Path

import scipy.stats as stats
from scipy.stats import beta, chisquare, kstest, ks_2samp, mannwhitneyu, anderson_ksamp, epps_singleton_2samp
from scipy.optimize import minimize
from scipy.interpolate import interp1d

import statsmodels.api as sm
from statsmodels.discrete.discrete_model import NegativeBinomial
from patsy import dmatrix, build_design_matrices
from pygam import PoissonGAM, s as gam_s

import wrds
WRDS_LOGIN = 'cssmyth'
```

```python
wrds_names_query = pd.read_parquet("data/wrds_names_query.parquet")
msf = pd.read_parquet("data/msf.parquet")
```

```python
# compute weight for all CRSP securities per month
msf = msf.copy()

msf['mktcap'] = msf['prc'].abs() * msf['shrout']
msf['total_mktcap'] = msf.groupby('date')['mktcap'].transform('sum')
msf['weight'] = msf['mktcap'] / msf['total_mktcap']

msf['pctile'] = (
    msf.groupby('date')['weight'].rank(pct=True, ascending=False)
)
```

```python
# make date a datetime type

msf['date'] = pd.to_datetime(msf['date'])

# start of security data and end of security data
wrds_names_query['securitybegdt'] = pd.to_datetime(wrds_names_query['securitybegdt'])
wrds_names_query['securityenddt'] = pd.to_datetime(wrds_names_query['securityenddt'])

# turn dates into month period for joins
msf['month']  = msf['date'].dt.to_period('M')
wrds_names_query['beg_month'] = wrds_names_query['securitybegdt'].dt.to_period('M')
wrds_names_query['end_month'] = wrds_names_query['securityenddt'].dt.to_period('M')
```

```python
# first and last month in msf universe
first_month = msf['month'].min()
last_month  = msf['month'].max()

drop_months = [
    pd.Period('1962-07', 'M'),
    pd.Period('1972-12', 'M'),
    pd.Period('2006-03', 'M'),
]

ent_keys = wrds_names_query.loc[
    wrds_names_query['beg_month'].notna(),
    ['permno','beg_month']
].copy()

ent_keys['next_month'] = ent_keys['beg_month'] + 1 

# entrances 
df_ent = pd.merge(
    msf[['permno','month','pctile','weight','mktcap','total_mktcap']],
    ent_keys,
    left_on  = ['permno','month'],
    right_on = ['permno','next_month'],
    how      = 'inner'
)

df_ent['event'] = 'entrance'
df_ent = df_ent[df_ent['month'] != first_month]
df_ent = df_ent[~df_ent['month'].isin(drop_months)]
df_ent = df_ent.assign(event_month=df_ent['month'])

exit_keys = wrds_names_query.loc[
    wrds_names_query['end_month'].notna(),
    ['permno','end_month']
].copy()

exit_keys['prev_month'] = exit_keys['end_month'] - 1  # Period[M] arithmetic


# exits 
df_exit = pd.merge(
    msf[['permno','month','pctile','weight','mktcap','total_mktcap']],
    exit_keys, 
    left_on  = ['permno','month'],
    right_on = ['permno','prev_month'],
    how      = 'inner'
)

df_exit['event'] = 'exit'
df_exit = df_exit[df_exit['month'] != last_month]
df_exit = df_exit[~df_exit['month'].isin(drop_months)]

df_exit = (df_exit
           .rename(columns={'end_month':'event_month'})
           .drop(columns=['prev_month'])
          )

# collapse to one entrance (earliest) and exit (latest) per permno
df_ent = (df_ent
          .sort_values('month')                    
          .drop_duplicates(subset='permno',     
                           keep='first')          
         )
df_exit = (df_exit
           .sort_values('month')                   
           .drop_duplicates(subset='permno',     
                            keep='last')          
          )

# combine
df_events = pd.concat([df_ent, df_exit], ignore_index=True)

# rename for downstream consistency
df_events.rename(columns={'pctile':'quantile'}, inplace=True)
```

```python
# counts of events by month and type
Y_increments = (
    df_events
    .groupby(['event_month', 'event'])
    .size()
    .unstack(fill_value=0)
    .rename(columns={'entrance': 'Y1_inc', 'exit': 'Y2_inc'})
    .sort_index()
)

# Y1 being num of entrances in that month, Y2 being num of exits in that month

Y_increments
```

```python
# total number of active securities per month

N_series = (
    msf
    .groupby('month')['permno']
    .nunique()
    .rename('N')
    .sort_index()
)

N_series.head()
```

```python
# turn increments into cumulative counting processes
Y_cum = (
    Y_increments
    .cumsum()
    .rename(columns={'Y1_inc': 'Y1_cum', 'Y2_inc': 'Y2_cum'})
)

Y_cum
```

```python
# combine everything into clean time series table

entrance_exit_panel = (
    N_series.to_frame()     
    .join(Y_increments, how='left')
    .fillna(0)                   
    .assign(
        Y1_cum=lambda df: df['Y1_inc'].cumsum(),
        Y2_cum=lambda df: df['Y2_inc'].cumsum(),
    )
    .sort_index()
)

entrance_exit_panel.head()
```

```python
# fit beta distributions with data adjustment
def fit_beta_distribution(data, event_type):
    """
    Fit beta distribution to quantile data
    Returns parameters and goodness-of-fit statistics
    """
    # handle boundary values
    clean_data = data.dropna()
    #clean_data = data
    
    # Beta distribution requires values strictly between 0 and 1
    # apply small adjustment to boundary values
    epsilon = 1e-6
    clean_data = clean_data.copy()
    clean_data[clean_data <= 0] = epsilon
    clean_data[clean_data >= 1] = 1 - epsilon
    
    print(f"\n{event_type} Data Summary:")
    print(f"Original range: [{data.min():.6f}, {data.max():.6f}]")
    print(f"Adjusted range: [{clean_data.min():.6f}, {clean_data.max():.6f}]")
    print(f"Number of boundary adjustments: {((data <= 0) | (data >= 1)).sum()}")

    # fit beta distribution using method of moments and MLE
    alpha_hat, beta_hat, loc, scale = stats.beta.fit(clean_data, floc=0, fscale=1)
    
    # calc goodness of fit using adjusted data
    ks_stat, ks_pvalue = stats.kstest(clean_data, lambda x: stats.beta.cdf(x, alpha_hat, beta_hat))
    
    # calc aic, bic
    log_likelihood = np.sum(stats.beta.logpdf(clean_data, alpha_hat, beta_hat))
    n = len(clean_data)
    aic = 2 * 2 - 2 * log_likelihood  # 2 parameters (alpha, beta)
    bic = np.log(n) * 2 - 2 * log_likelihood
    
    results = {
        'alpha': alpha_hat,
        'beta': beta_hat,
        'ks_statistic': ks_stat,
        'ks_pvalue': ks_pvalue,
        'aic': aic,
        'bic': bic,
        'log_likelihood': log_likelihood,
        'n_observations': n,
        'adjusted_data': clean_data  # Store the adjusted data for later use
    }

    print(f"\n{event_type} Beta Distribution Parameters:")
    print(f"Alpha: {alpha_hat:.4f}")
    print(f"Beta: {beta_hat:.4f}")
    print(f"KS Test p-value: {ks_pvalue:.4f}")
    print(f"AIC: {aic:.2f}")
    print(f"BIC: {bic:.2f}")
    
    return results

# fit distributions
entrance_params = fit_beta_distribution(ent_q, "ENTRANCE")
exit_params = fit_beta_distribution(exit_q, "EXIT")
```

```python
# calculate probabilities by market rank deciles
def calculate_entry_exit_probabilities(entrance_params, exit_params, quantile_range=None):
    if quantile_range is None:
        quantile_range = np.linspace(0, 1, 11)  # Deciles
    
    results = []
    
    for i in range(len(quantile_range)-1):
        lower = quantile_range[i]
        upper = quantile_range[i+1]
        
        # calculate probabilities using CDF
        entrance_prob = (stats.beta.cdf(upper, entrance_params['alpha'], entrance_params['beta']) - 
                        stats.beta.cdf(lower, entrance_params['alpha'], entrance_params['beta']))
        
        exit_prob = (stats.beta.cdf(upper, exit_params['alpha'], exit_params['beta']) - 
                    stats.beta.cdf(lower, exit_params['alpha'], exit_params['beta']))
        
        results.append({
            'quantile_range': f"{lower:.1f}-{upper:.1f}",
            'entrance_probability': entrance_prob,
            'exit_probability': exit_prob,
            'ratio_exit_to_entrance': exit_prob / entrance_prob if entrance_prob > 0 else np.inf
        })
    
    return pd.DataFrame(results)

# calculate probabilities
prob_df = calculate_entry_exit_probabilities(entrance_params, exit_params)
print("PROBABILITY ANALYSIS BY MARKET RANK DECILE:")
print(prob_df.round(4))
```

```python
# visualize probability analysis
fig, ax = plt.subplots(figsize=(12, 6))

# Probability comparison by decile
deciles = range(len(prob_df))
width = 0.35

ax.bar([x - width/2 for x in deciles], prob_df['entrance_probability'], 
           width, label='Entrance Probability', alpha=0.7, color='blue')
ax.bar([x + width/2 for x in deciles], prob_df['exit_probability'], 
           width, label='Exit Probability', alpha=0.7, color='red')
ax.set_xlabel('Market Rank Decile')
ax.set_ylabel('Probability')
ax.set_title('Entry vs Exit Probabilities by Market Rank Decile')
ax.set_xticks(deciles)
ax.set_xticklabels(prob_df['quantile_range'], rotation=45)
ax.legend()
ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
α_b, β_b = entrance_params['alpha'], entrance_params['beta']
α_d, β_d = exit_params['alpha'],     exit_params['beta']

# entrance/exit ratio
num_entrances = df_ent.dropna().shape[0]
num_exits = df_exit.dropna().shape[0]
ratio = num_entrances / num_exits if num_exits > 0 else np.inf

death_occurrence = 1.0
birth_occurrence = ratio * death_occurrence
print(f"birth_occurrence = {birth_occurrence:.6f}, death_occurrence = {death_occurrence:.6f}")

# rank-dependent intensities using Beta PDFs (clamped to avoid boundary NaNs)
def safe_pdf(x, α, β, eps=1e-6):
    x = np.clip(x, eps, 1-eps)
    return beta.pdf(x, α, β, loc=0, scale=1)

pdf_birth = lambda x: safe_pdf(x, α_b, β_b)
pdf_death = lambda x: safe_pdf(x, α_d, β_d)
λ_birth   = lambda x: birth_occurrence * pdf_birth(x)
λ_death   = lambda x: death_occurrence * pdf_death(x)
λ_total   = lambda x: λ_birth(x) + λ_death(x)

# seed particles from empirical percentiles at first month
raw = msf.loc[msf['month'] == first_month, 'pctile']
N0 = 20000
particles = np.random.choice(raw.dropna().values, size=N0, replace=True)
print("Seeded particle count:", particles.size)
```

```python
# branching particle simulation (Beta intensities, 10000 events)
t = 0.0
events = []

for i in range(10000):
    rates = λ_total(particles)
    taus = np.random.exponential(1.0 / rates)

    i_star = np.argmin(taus)
    dt = taus[i_star]
    t += dt
    x_i = particles[i_star]

    if np.random.rand() < λ_birth(x_i) / λ_total(x_i):
        particles = np.append(particles, x_i)
        particles = np.clip(particles, 1e-6, 1-1e-6)
        events.append(('birth', t, x_i))
    else:
        events.append(('death', t, x_i))
        particles = np.delete(particles, i_star)
```

```python
# simulation summary

df_sim = pd.DataFrame(events, columns=['event', 'time', 'percentile'])
counts = df_sim['event'].value_counts()
print("Total events:", len(df_sim))
print("Birth events:", counts.get('birth', 0))
print("Death events:", counts.get('death', 0))
```

```python
# birth and death density vs Beta PDF

x_vals = np.linspace(0, 1, 300)

plt.figure()
plt.hist(df_sim[df_sim['event']=='birth']['percentile'], bins=20, density=True, alpha=0.3)
plt.plot(x_vals, pdf_birth(x_vals))
plt.title('Simulated Birth Percentiles vs Beta PDF')
plt.xlabel('Percentile')
plt.ylabel('Density')
plt.savefig(out_dir / "birth_density_beta.pdf", bbox_inches="tight")
plt.show()

death = df_sim.loc[df_sim['event']=='death', 'percentile'].dropna()
pad = 0.01
lo = max(0.0, death.min() - pad)
hi = min(1.0, death.max() + pad)
F = lambda x: beta.cdf(x, α_d, β_d, loc=0, scale=1)
f = lambda x: beta.pdf(x, α_d, β_d, loc=0, scale=1)
Z = max(F(hi) - F(lo), 1e-12)
f_cond = lambda x: f(x) / Z

x = np.linspace(lo, hi, 800)
plt.figure()
plt.hist(death, bins=20, density=True, alpha=0.3)
plt.plot(x, f_cond(x))
plt.xlim(lo, hi)
plt.title('Simulated Death Percentiles vs Beta PDF')
plt.xlabel('Percentile')
plt.ylabel('Density')
plt.savefig(out_dir / "death_density_beta.pdf", bbox_inches="tight")
plt.show()
```

```python
# one-sample GOF: Beta simulation vs fitted Beta
sim_birth = df_sim.loc[df_sim['event']=='birth', 'percentile'].to_numpy()
sim_death = df_sim.loc[df_sim['event']=='death', 'percentile'].to_numpy()

def one_sample_gof_beta(sample, alpha, beta, label, m_bins=10, seed=0):
    """One-sample GOF vs Beta(alpha, beta):
       KS, Cramér–von Mises, binned chi-square (equal-probability bins), Wasserstein-1."""
    sample = np.asarray(sample)
    sample = sample[(sample > 0) & (sample < 1)]
    out = {'n': sample.size}
    if sample.size < 10:
        out['note'] = f'[{label}] too few samples for reliable GOF'
        return out

    ks  = stats.kstest(sample, 'beta', args=(alpha, beta))
    cvm = stats.cramervonmises(sample, stats.beta.cdf, args=(alpha, beta))

    q = np.linspace(0, 1, m_bins + 1)
    bins = stats.beta.ppf(q, alpha, beta)
    bins[0], bins[-1] = 0.0, 1.0
    hist, _ = np.histogram(sample, bins=bins)
    expected = np.full_like(hist, sample.size / m_bins, dtype=float)
    chi2 = stats.chisquare(hist, f_exp=expected)

    rng = np.random.default_rng(seed)
    ref = stats.beta.rvs(alpha, beta, size=min(5000, max(1000, 2*sample.size)), random_state=rng)
    w1 = stats.wasserstein_distance(sample, ref)

    return {
        'n': sample.size,
        'ks_stat': ks.statistic,   'ks_p': ks.pvalue,
        'cvm_stat': cvm.statistic, 'cvm_p': cvm.pvalue,
        'chi2_stat': chi2.statistic, 'chi2_p': chi2.pvalue,
        'w1': w1
    }

print('One-sample GOF vs target Beta')
birth_gof = one_sample_gof_beta(sim_birth, α_b, β_b, 'birth')
death_gof = one_sample_gof_beta(sim_death, α_d, β_d, 'death')
print('Birth:', birth_gof)
print('Death:', death_gof)

# Bernoulli calibration test for birth vs death choices
xj = df_sim['percentile'].to_numpy()
pj = stats.beta.pdf(xj, α_b, β_b)
qj = stats.beta.pdf(xj, α_d, β_d)
p_hat = (birth_occurrence * pj) / (birth_occurrence * pj + death_occurrence * qj)
y = (df_sim['event'].to_numpy() == 'birth').astype(float)

S = np.sum(y - p_hat)
V = np.sum(p_hat * (1.0 - p_hat))
z = S / np.sqrt(V) if V > 0 else np.nan
pval_cal = 2 * stats.norm.sf(np.abs(z))
print(f'\nBernoulli calibration: Z = {z:.3f}, p = {pval_cal:.4g}')
```

```python
# bin rank quantiles into equal-width percentile bins and count events
n_bins    = 100
bin_edges = np.linspace(0, 1, n_bins + 1)
bin_mids  = 0.5 * (bin_edges[:-1] + bin_edges[1:])
bin_width = 1.0 / n_bins

ent_clean  = ent_q.dropna().values
exit_clean = exit_q.dropna().values

ent_counts_bin,  _ = np.histogram(ent_clean,  bins=bin_edges)
exit_counts_bin, _ = np.histogram(exit_clean, bins=bin_edges)

N_ent  = len(ent_clean)
N_exit = len(exit_clean)

# offset = log(N_total * bin_width): expected log-count under uniform intensity
# constant across bins, so it only shifts the intercept, not the shape
log_offset_ent  = np.full(n_bins, np.log(N_ent  * bin_width))
log_offset_exit = np.full(n_bins, np.log(N_exit * bin_width))

print(f"Bins: {n_bins}  |  Entrance events: {N_ent}  |  Exit events: {N_exit}")
print(f"Expected count/bin (uniform): {N_ent * bin_width:.1f}")
print(f"Observed zero-count bins — Entrance: {(ent_counts_bin == 0).sum()}  |  Exit: {(exit_counts_bin == 0).sum()}")
print()
print(pd.DataFrame({
    'p': bin_mids, 'ent_count': ent_counts_bin, 'exit_count': exit_counts_bin
}).describe().round(1))
```

```python
# natural cubic spline basis via patsy
# save the design_info so we can apply the SAME knot scheme to arbitrary new p values
df_spline = 5  # degrees of freedom — smooth enough to capture the shape without overfitting

X_basis_info = dmatrix(f"cr(p, df={df_spline})", {'p': bin_mids}, return_type='matrix')
X_train = np.asarray(X_basis_info)

def predict_basis(p_new):
    """Apply training knot scheme to new p values (avoids knot extrapolation artifacts)."""
    p_new = np.atleast_1d(np.asarray(p_new, dtype=float))
    return np.asarray(
        build_design_matrices([X_basis_info.design_info], {'p': p_new})[0]
    )

# fine grid for plotting fitted intensity curves
p_grid = np.linspace(0.005, 0.995, 500)
X_grid = predict_basis(p_grid)

print(f"Design matrix: {X_train.shape[0]} bins × {X_train.shape[1]} basis functions  (df_spline={df_spline})")
```

```python
# Poisson GLM with natural spline basis
# Pearson chi2/df > 1 signals overdispersion → motivates Negative Binomial
pois_ent = sm.GLM(
    ent_counts_bin, X_train,
    family=sm.families.Poisson(),
    offset=log_offset_ent
).fit()

pois_exit = sm.GLM(
    exit_counts_bin, X_train,
    family=sm.families.Poisson(),
    offset=log_offset_exit
).fit()

print("Poisson GLM — Entrance")
print(f"  AIC: {pois_ent.aic:.1f}  |  Deviance: {pois_ent.deviance:.1f}  |  df_resid: {int(pois_ent.df_resid)}")
print(f"  Pearson χ²/df: {pois_ent.pearson_chi2 / pois_ent.df_resid:.2f}  (>1 signals overdispersion)")
print("Poisson GLM — Exit")
print(f"  AIC: {pois_exit.aic:.1f}  |  Deviance: {pois_exit.deviance:.1f}  |  df_resid: {int(pois_exit.df_resid)}")
print(f"  Pearson χ²/df: {pois_exit.pearson_chi2 / pois_exit.df_resid:.2f}")

# intensity shape on the fine grid (drop offset → get shape of lambda(p))
lam_pois_ent  = np.exp(X_grid @ pois_ent.params)
lam_pois_exit = np.exp(X_grid @ pois_exit.params)
```

```python
# Negative Binomial GLM — same spline basis, jointly estimates dispersion parameter alpha
# variance = mean + alpha * mean^2  (nb2 parameterization)
# alpha > 0 handles overdispersion from event clustering at top percentiles

nb_ent = NegativeBinomial(
    ent_counts_bin, X_train,
    exposure=np.exp(log_offset_ent),
    loglike_method='nb2'
).fit(disp=False)

nb_exit = NegativeBinomial(
    exit_counts_bin, X_train,
    exposure=np.exp(log_offset_exit),
    loglike_method='nb2'
).fit(disp=False)

# params = [β_0, ..., β_k, ln_alpha]; regression coefficients are params[:-1]
alpha_ent  = np.exp(nb_ent.params[-1])
alpha_exit = np.exp(nb_exit.params[-1])

print("Negative Binomial — Entrance")
print(f"  AIC: {nb_ent.aic:.1f}  |  dispersion α: {alpha_ent:.4f}")
print("Negative Binomial — Exit")
print(f"  AIC: {nb_exit.aic:.1f}  |  dispersion α: {alpha_exit:.4f}")
print("\n  (α ≈ 0 → near-Poisson; α > 0.1 → notable overdispersion, NegBin preferred)")

# intensity shape on the grid (without exposure → shape of λ(p))
lam_nb_ent  = np.exp(X_grid @ nb_ent.params[:-1])
lam_nb_exit = np.exp(X_grid @ nb_exit.params[:-1])
```

```python
# Poisson GAM via pygam — penalty-based smooth, no need to pre-specify knot locations
gam_ent  = PoissonGAM(gam_s(0, n_splines=20)).fit(bin_mids.reshape(-1, 1), ent_counts_bin)
gam_exit = PoissonGAM(gam_s(0, n_splines=20)).fit(bin_mids.reshape(-1, 1), exit_counts_bin)

print(f"GAM Entrance — AIC: {gam_ent.statistics_['AIC']:.1f}  |  eff. df: {gam_ent.statistics_['edof']:.1f}")
print(f"GAM Exit     — AIC: {gam_exit.statistics_['AIC']:.1f}  |  eff. df: {gam_exit.statistics_['edof']:.1f}")

# predict returns expected counts on the same scale as training data
# since all bins have the same offset (constant), the shape is comparable to the GLM curves
lam_gam_ent  = gam_ent.predict(p_grid.reshape(-1, 1))
lam_gam_exit = gam_exit.predict(p_grid.reshape(-1, 1))
```

```python
# normalize all curves to probability density (integrate to 1 over (0,1)) for comparison
def to_density(lam, p):
    area = np.trapezoid(lam, p)
    return lam / area if area > 0 else lam

x_beta = np.linspace(0.001, 0.999, 500)
lam_beta_ent_grid  = stats.beta.pdf(x_beta, α_b, β_b)
lam_beta_exit_grid = stats.beta.pdf(x_beta, α_d, β_d)

curves = {
    'Entrance': {
        'data':    ent_clean,
        'Beta':    (x_beta, to_density(lam_beta_ent_grid,  x_beta)),
        'Poisson': (p_grid,  to_density(lam_pois_ent,       p_grid)),
        'NegBin':  (p_grid,  to_density(lam_nb_ent,         p_grid)),
        'GAM':     (p_grid,  to_density(lam_gam_ent,        p_grid)),
    },
    'Exit': {
        'data':    exit_clean,
        'Beta':    (x_beta, to_density(lam_beta_exit_grid, x_beta)),
        'Poisson': (p_grid,  to_density(lam_pois_exit,      p_grid)),
        'NegBin':  (p_grid,  to_density(lam_nb_exit,        p_grid)),
        'GAM':     (p_grid,  to_density(lam_gam_exit,       p_grid)),
    },
}

fig, axes = plt.subplots(1, 2, figsize=(13, 5))
styles = {'Beta': ('-', 'C0'), 'Poisson': ('--', 'C1'), 'NegBin': ('-.', 'C2'), 'GAM': (':', 'C3')}

for ax, (title, d) in zip(axes, curves.items()):
    ax.hist(d['data'], bins=50, density=True, alpha=0.2, color='gray', label='Data')
    for mname, val in d.items():
        if mname == 'data':
            continue
        px, py = val
        ls, col = styles[mname]
        ax.plot(px, py, ls=ls, color=col, lw=2, label=mname)
    ax.set_xlabel('Market Rank Quantile')
    ax.set_ylabel('Density')
    ax.set_title(title)
    ax.legend(fontsize=9)
    ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig(out_dir / "glm_gam_comparison.pdf", bbox_inches='tight')
plt.show()

# boundedness table: the key result for the theorem
print("\n=== Boundedness of intensity density at tail (max over p ∈ (0,1)) ===")
print(f"{'Model':<18} {'Entrance max':>14} {'Exit max':>14}")
print("-" * 50)
for mname in ['Beta', 'Poisson', 'NegBin', 'GAM']:
    _, py_ent = curves['Entrance'][mname]
    _, py_ex  = curves['Exit'][mname]
    print(f"{mname:<18} {py_ent.max():>14.4f} {py_ex.max():>14.4f}")
print("\n(Beta density diverges as p→1 for α<1 or β<1; GLM/GAM intensities are bounded by construction)")
```

```python
# build callable λ_birth(x) and λ_death(x) from NegBin fit (normalized to density)
# these replace the Beta PDFs in the particle simulation

_coef_birth = nb_ent.params[:-1]
_coef_death = nb_exit.params[:-1]

# normalization constants: integrate shape over interior grid to avoid boundary artifacts
_p_norm = np.linspace(0.005, 0.995, 2000)
_X_norm = predict_basis(_p_norm)
_norm_birth = np.trapezoid(np.exp(_X_norm @ _coef_birth), _p_norm)
_norm_death = np.trapezoid(np.exp(_X_norm @ _coef_death), _p_norm)

def λ_birth_nb(x):
    x = np.atleast_1d(np.asarray(x, dtype=float))
    x = np.clip(x, 0.005, 0.995)
    return birth_occurrence * np.exp(predict_basis(x) @ _coef_birth) / _norm_birth

def λ_death_nb(x):
    x = np.atleast_1d(np.asarray(x, dtype=float))
    x = np.clip(x, 0.005, 0.995)
    return death_occurrence * np.exp(predict_basis(x) @ _coef_death) / _norm_death

def λ_total_nb(x):
    return λ_birth_nb(x) + λ_death_nb(x)

# sanity checks: compare NegBin vs Beta intensity at key percentiles
p_check = np.array([0.1, 0.5, 0.9, 0.99])
print(f"{'p':>5}  {'λ_birth_nb':>12} {'λ_death_nb':>12} {'λ_birth_beta':>14} {'λ_death_beta':>14}")
for p in p_check:
    lb_nb = λ_birth_nb(p)[0]
    ld_nb = λ_death_nb(p)[0]
    lb_b  = λ_birth(np.array([p]))[0]
    ld_b  = λ_death(np.array([p]))[0]
    print(f"{p:>5.2f}  {lb_nb:>12.4f} {ld_nb:>12.4f} {lb_b:>14.4f} {ld_b:>14.4f}")

print(f"\nλ_birth_nb integrates to: {np.trapezoid(λ_birth_nb(_p_norm), _p_norm) / birth_occurrence:.4f}  (target ≈ 1)")
print(f"λ_death_nb integrates to: {np.trapezoid(λ_death_nb(_p_norm), _p_norm) / death_occurrence:.4f}  (target ≈ 1)")
print("\nλ_birth_nb and λ_death_nb are now available as drop-in replacements for the Beta intensities.")
```

```python
# Branching particle simulation with NegBin intensities

raw_nb = msf.loc[msf['month'] == first_month, 'pctile']
clean_nb = raw_nb.dropna().values
N0_nb = 20000
particles_nb = np.random.choice(clean_nb, size=N0_nb, replace=True)
particles_nb = np.clip(particles_nb, 0.005, 0.995)
 
print(f"Seeded {particles_nb.size} particles for NegBin simulation")
 
t_nb = 0.0
events_nb = []
N_events_nb = 10000
 
for _ in range(N_events_nb):
    rates = λ_total_nb(particles_nb)
    taus = np.random.exponential(1.0 / rates)
 
    i_star = np.argmin(taus)
    dt = taus[i_star]
    t_nb += dt
    x_i = particles_nb[i_star]
 
    if np.random.rand() < λ_birth_nb(x_i) / λ_total_nb(x_i):
        particles_nb = np.append(particles_nb, x_i)
        particles_nb = np.clip(particles_nb, 0.005, 0.995)
        events_nb.append(('birth', t_nb, x_i))
    else:
        events_nb.append(('death', t_nb, x_i))
        particles_nb = np.delete(particles_nb, i_star)
 
df_sim_nb = pd.DataFrame(events_nb, columns=['event', 'time', 'percentile'])
counts_nb = df_sim_nb['event'].value_counts()
print(f"Total events: {len(df_sim_nb)}")
print(f"Birth events: {counts_nb.get('birth', 0)}")
print(f"Death events: {counts_nb.get('death', 0)}")
```

```python
# Consolidated goodness-of-fit comparison table
# Comparing Beta, Poisson GLM, NegBin GML, and GAM on same footing
 
# helper: compute binned chi-square of a fitted density against histogram counts
def binned_chi2(data, fitted_density_func, p_grid_func, n_test_bins=30):
    """Chi-square test: bin the data, compute expected counts from fitted density."""
    edges = np.linspace(0, 1, n_test_bins + 1)
    mids = 0.5 * (edges[:-1] + edges[1:])
    observed, _ = np.histogram(data, bins=edges)
    widths = np.diff(edges)
    density_at_mids = fitted_density_func(mids)
    expected = density_at_mids * widths * len(data)
    mask = expected >= 5
    if mask.sum() < 3:
        return np.nan, np.nan
    obs_m = observed[mask]
    exp_m = expected[mask]
    # renormalize expected to match observed sum (required by scipy.chisquare)
    exp_m = exp_m * (obs_m.sum() / exp_m.sum())
    chi2_stat, chi2_p = chisquare(obs_m, f_exp=exp_m)
    return chi2_stat, chi2_p
 
# helper: KS test against a fitted CDF (numerical integration of density)
def numerical_ks(data, density_func, p_grid, n_grid=2000):
    """KS test using numerically integrated CDF from density."""
    grid = np.linspace(0.001, 0.999, n_grid)
    density_vals = density_func(grid)
    cdf_vals = np.cumsum(density_vals) * (grid[1] - grid[0])
    cdf_vals = cdf_vals / cdf_vals[-1]  # normalize
    cdf_interp = interp1d(grid, cdf_vals, bounds_error=False, fill_value=(0, 1))
    ks_stat, ks_p = kstest(data, cdf_interp)
    return ks_stat, ks_p
 
# density functions for each model (normalized to integrate to 1)
def make_density(raw_func, p_grid):
    vals = raw_func(p_grid)
    area = np.trapezoid(vals, p_grid)
    return lambda x: raw_func(x) / area
 
p_fine = np.linspace(0.005, 0.995, 2000)
 
density_funcs = {
    'Beta': {
        'entrance': make_density(λ_birth, p_fine),
        'exit':     make_density(λ_death, p_fine),
    },
    'Poisson': {
        'entrance': make_density(lambda x: np.exp(predict_basis(x) @ pois_ent.params), p_fine),
        'exit':     make_density(lambda x: np.exp(predict_basis(x) @ pois_exit.params), p_fine),
    },
    'NegBin': {
        'entrance': make_density(lambda x: np.exp(predict_basis(x) @ nb_ent.params[:-1]), p_fine),
        'exit':     make_density(lambda x: np.exp(predict_basis(x) @ nb_exit.params[:-1]), p_fine),
    },
    'GAM': {
        'entrance': make_density(lambda x: gam_ent.predict(np.atleast_1d(x).reshape(-1, 1)), p_fine),
        'exit':     make_density(lambda x: gam_exit.predict(np.atleast_1d(x).reshape(-1, 1)), p_fine),
    },
}
 
# AIC values (Beta doesn't have a comparable AIC from the same framework,
# so we report log-likelihood of data under fitted Beta as a proxy)
beta_ll_ent  = np.sum(stats.beta.logpdf(np.clip(ent_clean, 0.001, 0.999), α_b, β_b))
beta_ll_exit = np.sum(stats.beta.logpdf(np.clip(exit_clean, 0.001, 0.999), α_d, β_d))
beta_aic_ent  = -2 * beta_ll_ent  + 2 * 2   # 2 parameters
beta_aic_exit = -2 * beta_ll_exit + 2 * 2
 
aic_table = {
    'Beta':    (beta_aic_ent,   beta_aic_exit),
    'Poisson': (pois_ent.aic,   pois_exit.aic),
    'NegBin':  (nb_ent.aic,     nb_exit.aic),
    'GAM':     (gam_ent.statistics_['AIC'], gam_exit.statistics_['AIC']),
}
 
# build results
rows = []
for model_name in ['Beta', 'Poisson', 'NegBin', 'GAM']:
    for event_type, data in [('Entrance', ent_clean), ('Exit', exit_clean)]:
        key = 'entrance' if event_type == 'Entrance' else 'exit'
        dfunc = density_funcs[model_name][key]
        chi2_stat, chi2_p = binned_chi2(data, dfunc, p_fine)
        ks_stat, ks_p = numerical_ks(data, dfunc, p_fine)
        aic_val = aic_table[model_name][0 if key == 'entrance' else 1]
 
        # check boundedness
        vals_at_boundary = dfunc(np.array([0.001, 0.999]))
        is_bounded = np.all(np.isfinite(vals_at_boundary)) and np.max(vals_at_boundary) < 100
 
        rows.append({
            'Model': model_name,
            'Event': event_type,
            'AIC': aic_val,
            'χ² stat': chi2_stat,
            'χ² p': chi2_p,
            'KS stat': ks_stat,
            'KS p': ks_p,
            'Bounded': is_bounded,
        })
 
gof_df = pd.DataFrame(rows)
print("=== Goodness-of-Fit Comparison ===\n")
print(gof_df.to_string(index=False, float_format='%.4f'))
print("\nKey takeaway: NegBin matches data quality of Beta while guaranteeing bounded intensities.")
```

```python
# One-sample goodness of fit of NegBin simulation
# Analogous to cell 43 for Beta (tests whether NegBin simulation output matches NegBin fitted density 
def one_sample_gof_nb(sample, density_func, label, m_bins=10, seed=0):
    """One-sample GOF vs a numerically specified density."""
    sample = np.asarray(sample)
    sample = sample[(sample > 0) & (sample < 1)]
    if sample.size < 10:
        print(f'[{label}] too few samples for reliable GOF')
        return {}
 
    # build CDF by numerical integration
    grid = np.linspace(0.001, 0.999, 5000)
    pdf_vals = density_func(grid)
    cdf_vals = np.cumsum(pdf_vals) * (grid[1] - grid[0])
    cdf_vals = cdf_vals / cdf_vals[-1]
    cdf_interp = interp1d(grid, cdf_vals, bounds_error=False, fill_value=(0, 1))
 
    # KS test
    ks = kstest(sample, cdf_interp)
 
    # binned chi-square with equal-probability bins
    q_edges = np.linspace(0, 1, m_bins + 1)
    ppf_vals = interp1d(cdf_vals, grid, bounds_error=False, fill_value=(0.001, 0.999))
    bin_edges = ppf_vals(q_edges)
    bin_edges[0], bin_edges[-1] = 0.0, 1.0
    hist, _ = np.histogram(sample, bins=bin_edges)
    expected = np.full_like(hist, sample.size / m_bins, dtype=float)
    chi2 = chisquare(hist, f_exp=expected)
 
    # Wasserstein distance vs synthetic sample from the fitted density
    rng = np.random.default_rng(seed)
    ref_u = rng.uniform(size=5000)
    ref_samples = ppf_vals(ref_u)
    w1 = stats.wasserstein_distance(sample, ref_samples)
 
    print(f"{label}: n={sample.size}")
    print(f"  KS: stat={ks.statistic:.4f}, p={ks.pvalue:.4g}")
    print(f"  χ²: stat={chi2.statistic:.2f}, p={chi2.pvalue:.4g}")
    print(f"  W1: {w1:.4f}")
    return {'ks_stat': ks.statistic, 'ks_p': ks.pvalue,
            'chi2_stat': chi2.statistic, 'chi2_p': chi2.pvalue, 'w1': w1}
 
print("=== One-sample GOF: NegBin simulation vs NegBin fitted density ===\n")
one_sample_gof_nb(sim_birth_nb.values, nb_birth_density_func := make_density(
    lambda x: np.exp(predict_basis(x) @ nb_ent.params[:-1]), p_fine), 'Birth (NegBin)')
print()
one_sample_gof_nb(sim_death_nb.values, nb_death_density_func := make_density(
    lambda x: np.exp(predict_basis(x) @ nb_exit.params[:-1]), p_fine), 'Death (NegBin)')
```
